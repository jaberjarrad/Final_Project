```python
import pandas as pd
```

```python
heartDisease = pd.read_csv("tubes2_HeartDisease_train.csv")
```

```python
heartDisease = heartDisease[heartDisease.Column4 != '?']
heartDisease = heartDisease[heartDisease.Column5 != '?']
heartDisease = heartDisease[heartDisease.Column6 != '?']
heartDisease = heartDisease[heartDisease.Column7 != '?']
heartDisease = heartDisease[heartDisease.Column8 != '?']
heartDisease = heartDisease[heartDisease.Column9 != '?']
heartDisease = heartDisease[heartDisease.Column10 != '?']
heartDisease = heartDisease[heartDisease.Column11 != '?']
heartDisease = heartDisease[heartDisease.Column12 != '?']
heartDisease = heartDisease[heartDisease.Column13 != '?']
```
